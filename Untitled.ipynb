```python
import pandas as pd
import json
import os
import codecs
import requests
```

```python
elastic_host = 'localhost'
```

```python
database = None
with open('database.json') as json_file:
    database = json.load(json_file)
```

```python
for d in database:
    del d['_id']
database
```

```python
response = requests.put(f"http://{elastic_host}:9200/fabrica-inteligente")
response.json()
```

```python
for item in database:
        response = requests.post(f"http://{elastic_host}:9200/fabrica-inteligente/_doc/", 
                                 data=json.dumps(item), 
                                 headers={
                                     'Content-Type': 'application/json'})
        print(response.json())
```
